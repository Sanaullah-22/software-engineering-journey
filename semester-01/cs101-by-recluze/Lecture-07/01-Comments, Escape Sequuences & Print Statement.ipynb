

                                                    ESCAPE SEQUENCES
        1. Scape Sequences are used to include special character in code
        Common scape sequence
            1 \n : New line
            2 \t : tab
            3 \\ : backslash
            4 \" : Double Quotes 
            5 \' : single quoute
    

```python
print("hey How are \n i am fine ") #new Line 
```

```python
print("hello g \t hello") #adding tab
```

```python
print("guys i want to add \\ backslash in my Code") # adding  backslash
```
